```python
import wbgapi as wb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_data(code):
    worldbank_data = wb.data.DataFrame(code, ['BRA'], range(1990, 2020, 1))
    return worldbank_data
```

```python
def show_corr(df):
    plt.figure(figsize=(16, 12))
    sns.heatmap(df, annot=True)
    plt.show()
```

## Agriculture

```python
agriculture = read_data(['NV.AGR.TOTL.CD','SL.AGR.EMPL.ZS','SP.RUR.TOTL.ZS','AG.PRD.FOOD.XD'])
```

```python
agriculture.fillna(agriculture.mean(),inplace=True)
```

```python
agriculture.loc[['SL.AGR.EMPL.ZS','SP.RUR.TOTL.ZS','AG.PRD.FOOD.XD']].sum()
```

```python
agriculture
```

```python
corr_agriculture = agriculture.T.corr(method='pearson')
```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(corr_agriculture, annot=True)
plt.show()
```
